```python
#data exploration
#importing the pandas library
import pandas as pd
```

```python
#read the .csv file and store it in the 'irizz' variable
irizz = pd.read_csv('Iris.csv')
#view first 10 dataset rows
print(irizz.head(10))
```

```python
irizz.shape
#the iris dataset has 150 rows and 6 columns
```

```python
desc = irizz.describe()
#data specifications
print(desc)
```

```python
info = irizz.info()
#dataset technical information
print(info)
```

```python
count = irizz.value_counts()
#shows how often the data reoccurs
print(count)
```

```python
#dataset variables, i.e a variable for every column
id = irizz["Id"]
sl = irizz["SepalLengthCm"]
sw = irizz["SepalWidthCm"]
pl = irizz["PetalLengthCm"]
pw = irizz["PetalWidthCm"]
sp = irizz["Species"]
```

```python
#importing the matplotlib and seaborn libraries and other necessary stuff
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
import seaborn as sb
```

```python
#ploting histograms using matplotlib
#Ploting a Sepal Length histogram
plt.hist(sl, bins = 25, edgecolor = 'black')
plt.title("Iris Sepal Lengths")
plt.xlabel("Sepal Lengths in cm")
plt.ylabel("No of Species")
```

```python
#Ploting a Sepal Width histogram
plt.hist(sw, bins = 25, edgecolor = 'black')
plt.title("Iris Sepal Widths")
plt.xlabel("Sepal Widths in cm")
plt.ylabel("No of Species")
```

```python
#Ploting a Petal Length histogram
plt.hist(pl, bins = 25, edgecolor = 'black')
plt.title("Iris Petal Lengths")
plt.xlabel("Petal Lengths in cm")
plt.ylabel("No of Species")
```

```python
#Ploting a Petal Width histogram
plt.hist(pw, bins = 25, edgecolor = 'black')
plt.title("Iris Petal Width")
plt.xlabel("Petal Widths in cm")
plt.ylabel("No of Species")
```

```python
#Ploting a Species histogram
plt.hist(sp, bins = 25, edgecolor = 'black')
plt.title("Iris Species")
plt.xlabel("Species")
plt.ylabel("No of Species")
```

```python
#plotting scatter plots using matplotlib
#sepal length and width
plt.scatter(sl, sw)
plt.title("Iris Sepal Dimensions")
plt.ylabel("Sepal Widths in cm")
plt.xlabel("Sepal Lengths in cm")
```

```python
#petal length and width
plt.scatter(pl, pw)
plt.title("Iris Petal Dimensions")
plt.ylabel("Petal Widths in cm")
plt.xlabel("Petal Lengths in cm")
```

```python
#petal length and sepal width
plt.scatter(pl, sw)
plt.title("Iris Petal and Sepal Relation")
plt.ylabel("Sepal Widths in cm")
plt.xlabel("Petal Lengths in cm")
```

```python
#sepal length and petal width
plt.scatter(sl, pw)
plt.title("Iris Petal and Sepal Relation")
plt.ylabel("Petal Widths in cm")
plt.xlabel("Sepal Lengths in cm")
```

```python
#sepal length and species
plt.scatter(sp, sl)
plt.title("Iris Sepal Length and Species Relation")
plt.xlabel("Species")
plt.ylabel("Sepal Lengths in cm")
```

```python
#sepal width and species
plt.scatter(sp, sw)
plt.title("Iris Sepal Width and Species Relation")
plt.xlabel("Species")
plt.ylabel("Sepal Widths in cm")
```

```python
#petal length and species
plt.scatter(sp, pl)
plt.title("Iris Petal Length and Species Relation")
plt.xlabel("Species")
plt.ylabel("PPetal Lengths in cm")
```

```python
#petal width and species
plt.scatter(sp, pw)
plt.title("Iris Petal Width and Species Relation")
plt.xlabel("Species")
plt.ylabel("Petal Widths in cm")
```
